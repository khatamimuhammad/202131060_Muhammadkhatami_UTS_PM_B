```python
#import beberapa library
import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt 
from sklearn.linear_model import LinearRegression 
```

```python
#membaca dataset dan memanggilnya
df = pd.read_csv ('R04_corruptions.csv') 
df
```

```python
#memanggil datasetnya 
df = pd.read_csv ('R04_corruptions.csv')  #y indepandent x dependent
df.columns = ['X','Y']
df
```

```python
df.corr()
```

```python
x_train = df['X'].values[:,np.newaxis]
y_train = df['Y'].values
```

```python
print(x_train)
```

```python
lm = LinearRegression()
lm.fit(x_train,y_train)
```

```python
print('Coefficient : ' + str(lm.coef_)) #mencetak koefisien regresi 
print('Intercept : ' + str(lm.intercept_)) #mencetak 

#mencetak koefisien kemiringan dari garis regresi
#menggambarkan hubungan ;inear antar variabel 'x' dan 'y' dalam model regresi
```

```python
#prepare plot
pb = lm.predict(x_train)
dfc = pd.DataFrame({'X': df['X'],'Y':pb})
plt.scatter(df['X'],df['Y'])
plt.plot(dfc['X'],dfc['Y'],color='green',linewidth=1)
plt.xlabel('Corruption Perception')
plt.ylabel('Government Transparency')
plt.show()
```
