```python
#import library
import pandas as pd
import numpy as np
```

```python
#membaca dataset format csv
data = pd.read_csv("K04_bank_customers.csv",sep=';')
```

```python
#panggil datasetnya
data
```

```python
#melihat colom apa saya yang ada dataset
data.info()
```

```python
#untuk mendapatkan ringkasan statistik deskriptif dari suatu DataFrame
data.describe()
```

```python
#mengubah beberapa kolom menjadi numerik
from sklearn.preprocessing import LabelEncoder

for col in ['job', 'marital', 'education', 'default', 'balance', 'housing', 'loan', 'contact', 'month', 'poutcome', 'y']:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    
print(data['job'].unique(),data['marital'].unique(),data['education'].unique(),data['default'].unique(),data['balance'].unique(),data['housing'].unique(),data['loan'].unique(),data['contact'].unique(),data['month'].unique(),data['poutcome'].unique(),data['y'].unique())
```

```python
#panggil datanya kembali
data
```

```python
#memisahkan variabel x dan y
x = data.drop(['y'], axis=1)
y = data['y']
```

```python
from sklearn.model_selection import train_test_split
```

```python
#training dan testing data tersebut dengan perbandingan 80:20
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
#memanggil algoritma NB
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

modelNB = GaussianNB()
modelNB.fit(x_train, y_train)
```

```python
hasil_prediksi = modelNB.predict(x_test)
```

```python
#melihar hasil akurasi model
from sklearn.metrics import accuracy_score
print('Akurasi Model: ', accuracy_score(y_test, hasil_prediksi))
```

```python
#membuat data baru dan melihat hasil prediksinya
x_new = np.array([[30, 9, 2, 3, 0, 3150, 1, 1, 1, 17, 10, 82, 3, -1, 3, 2]])
prediksi = modelNB.predict(x_new)
print(prediksi)
```
